# CIFAR-10

```python
VAR_CAT_DESCRIPTIONS["class_rel"] = (
    "Compares the effect of same-vs-different transformations with the effect of the class-relationship of the training task to target task: subset (2 out of 10), same (5), disjoint (different 5), superset (10)."
)
```

## Linear probing

```python
ci10_lp_figs = res_util.show_results(
    dataset_type="cifar10",
    variation_categories=["class_rel", "arch", "pt_samples", "ft_samples"],
    variation_descriptions=VAR_CAT_DESCRIPTIONS,
    ft_type="lp",
    seeds=list(range(10)),
    object_cardinality="default",
    show_legend_for={"class_rel"},
)
```

```python
## Save figures
PLOTS_DIR = Path("../figures/how_important/tvo_cifar")
for key, fig in ci10_lp_figs.items():
    fig.savefig(PLOTS_DIR / f"cifar10_{key}.pdf")
```
