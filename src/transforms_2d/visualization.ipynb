```python
import transforms_2d as t2d
from lib_vision.visualization.images import show_dataset_samples
```

## Dataset examples

```python
n_samples = 8
img_size = 64

config = t2d.Transforms2DConfig(
    sampling_seed=483,
    img_size=64,
    n_training_samples=n_samples,
    n_val_samples=n_samples,
    n_test_samples=n_samples,
    batch_size=n_samples,
)
for i, transform in enumerate(t2d.TRANSFORMS):
    print(f"{transform} transform")
    dataset = t2d.create_t2d_dataset(
        config,
        transforms=[transform],
        foregrounds=[i],
        # Disable normalization to 0 mean and 1 std dev for visualization.
        # Enable normalization for data that is consumed by a model.
        normalize=False,
    )
    fig = show_dataset_samples(
        dataset,
        n_samples=n_samples,
        data_type="test",
        figsize=(8, 6),
    )
```
